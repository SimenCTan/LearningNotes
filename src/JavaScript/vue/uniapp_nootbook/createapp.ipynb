### [创建uniapp项目](https://uniapp.dcloud.net.cn/quickstart-cli.html#%E8%BF%90%E8%A1%8C%E3%80%81%E5%8F%91%E5%B8%83uni-app)
创建 UniApp 项目可以通过以下几种方式：

### 1. 使用 HBuilderX

HBuilderX 是 DCloud 官方提供的开发工具，支持快速创建和管理 UniApp 项目。

**步骤：**

1. 下载并安装 HBuilderX。
2. 打开 HBuilderX，选择“文件” -> “新建” -> “项目”。
3. 在弹出的对话框中选择“UniApp”模板。
4. 填写项目名称、路径等信息，然后点击“创建”。
5. 项目创建完成后，可以在 HBuilderX 中进行开发和调试。

### 2. 使用 Vue CLI

如果你更习惯使用命令行，可以通过 Vue CLI 创建 UniApp 项目。

**步骤：**

1. 确保你已经安装了 Node.js 和 npm。
2. 全局安装 `@vue/cli`：
   ```bash
   npm install -g @vue/cli
   ```
3. 创建 UniApp 项目：
   ```bash
   vue create -p dcloudio/uni-preset-vue my-project
   ```
   这里 `my-project` 是你的项目名称。
4. 进入项目目录：
   ```bash
   cd my-project
   ```
5. 启动项目：
   ```bash
   npm run dev:%PLATFORM%
   ```
   其中 `%PLATFORM%` 可以替换为 `h5`、`mp-weixin` 等目标平台。
6. 使用alpha版本
    ```bash
    vue create -p dcloudio/uni-preset-vue#alpha my-alpha-project
    ```

7. 使用Vue3/Vite版
   ```bash
   npx degit dcloudio/uni-preset-vue#vite my-vue3-project
   npx degit dcloudio/uni-preset-vue#vite-alpha my-vue3-project
   npx degit dcloudio/uni-preset-vue#vite-ts my-vue3-project
   ```

### 3. 使用 CLI 工具

DCloud 还提供了一个命令行工具，可以通过它来创建 UniApp 项目。

**步骤：**

1. 安装 `@dcloudio/uni-cli`：
   ```bash
   npm install -g @dcloudio/uni-cli
   ```
2. 创建项目：
   ```bash
   uni create my-project
   ```
3. 进入项目目录并安装依赖：
   ```bash
   cd my-project
   npm install
   ```
4. 启动项目：
   ```bash
   npm run dev
   ```
5. 更新模版
   npx @dcloudio/uvm@latest

### 总结

以上是创建 UniApp 项目的几种常用方式。根据你的开发习惯和需求选择适合你的方式即可。

nvm（Node Version Manager）是一个用于管理 Node.js 版本的工具。以下是 nvm 的一些基本命令和用法：

### 1. 安装 nvm

如果你还没有安装 nvm，可以按照以下步骤进行安装：

#### macOS 和 Linux

在终端中运行：

```bash
curl -o- https://raw.githubusercontent.com/nvm-sh/nvm/v0.39.5/install.sh | bash
```

或者使用 wget：

```bash
wget -qO- https://raw.githubusercontent.com/nvm-sh/nvm/v0.39.5/install.sh | bash
```

安装后，确保在你的 shell 配置文件（如 `~/.bashrc` 或 `~/.zshrc`）中添加以下内容，以便在每次打开终端时加载 nvm：

```bash
export NVM_DIR="$HOME/.nvm"
[ -s "$NVM_DIR/nvm.sh" ] && \. "$NVM_DIR/nvm.sh" # This loads nvm
```

#### Windows

在 Windows 上，可以使用 [nvm-windows](https://github.com/coreybutler/nvm-windows)：

1. 下载并运行安装程序。
2. 按照安装向导进行操作。

### 2. 常用命令

- **查看可用 Node.js 版本**：

  ```bash
  nvm ls-remote
  ```

- **安装 Node.js 版本**：

  安装最新的 LTS 版本：

  ```bash
  nvm install --lts
  ```

  安装特定版本：

  ```bash
  nvm install 14.17.0
  ```

- **使用 Node.js 版本**：

  切换到已安装的版本：

  ```bash
  nvm use 14.17.0
  ```

- **查看已安装的版本**：

  ```bash
  nvm ls
  ```

- **设置默认版本**：

  设置默认的 Node.js 版本：

  ```bash
  nvm alias default 14.17.0
  ```

- **卸载 Node.js 版本**：

  卸载特定版本：

  ```bash
  nvm uninstall 14.17.0
  ```

### 3. 示例

假设你想要安装并使用 Node.js 16.x 版本，你可以这样做：

```bash
nvm install 16
nvm use 16
```

### 4. 检查安装

安装完成后，可以通过以下命令检查当前使用的 Node.js 版本：

```bash
node -v
```

### 5. 其他

nvm 还支持其他功能，如安装 npm、查看当前使用的 Node.js 版本等。你可以通过运行 `nvm help` 查看所有可用命令。

### uniapp和原生小程序开发的区别
根据你提供的图片，uni-app 和原生小程序开发的区别主要体现在以下几个方面：

1. **文件结构**：
   - 每个页面都是一个 `.vue` 文件，符合 Vue.js 的规范。开发者可以使用 Vue.js 的特性和语法。

2. **属性绑定**：
   - 在原生小程序中，属性绑定通常使用 `src="{{ url }}"` 的方式，而在 uni-app 中则升级为 `:src="url"`，这符合 Vue 的语法。

3. **事件绑定**：
   - 原生小程序使用 `bindtap="eventName"` 来绑定事件，而在 uni-app 中则使用 `@tap="eventName"`，支持传参的方式。

4. **Vue 指令支持**：
   - uni-app 支持常用的 Vue 指令，如 `v-for`、`v-if`、`v-show` 和 `v-model` 等，这使得开发更加灵活和方便。

5. **接口能力**：
   - 提示中建议使用前缀 wx 来替换为 uni，以便更好地适应多端开发的习惯。

总结来说，uni-app 在开发中更贴近 Vue.js 的语法和特性，使得开发者能够更灵活地进行多端应用的开发，而原生小程序则有其特定的语法和限制。

### 用vscode开发 uniapp
使用 VS Code 开发 uni-app 项目可以按照以下步骤进行：

1. **安装 uni-app 插件**：
   - 在 VS Code 中搜索并安装 `uni-app` 相关插件，例如 `uni-helper` 和 `uni-create-view`，以便获得代码提示和项目模板。

2. **快速创建页面**：
   - 使用安装的插件可以快速创建 uni-app 页面，提升开发效率。

3. **类型检查**：
   - 如果使用 TypeScript，确保安装类型声明文件。使用命令：
     ```bash
     pnpm i -D @types/wechat-miniprogram @uni-helper/uni-app-types
     ```

4. **配置 TypeScript**：
   - 配置 `tsconfig.json` 文件，添加必要的类型信息。这将帮助 VS Code 正确识别和提示代码。

5. **解决 JSON 关联问题**：
   - 在 VS Code 中设置文件关联，以确保 `.json` 文件能够被正确识别和处理。例如，可以将 `manifest.json` 和 `pages.json` 文件与特定的语言模式关联。

6. **使用代码提示**：
   - 在编写代码时，利用插件提供的代码提示功能，提升开发效率。

通过以上步骤，你可以在 VS Code 中高效地开发 uni-app 项目，享受更好的开发体验。
