- **请求工具**：通过拦截器，可以对请求进行统一处理，比如添加基础地址、超时时间、请求头标识和token等。

### 2. 使用方法
- **添加拦截器**：使用`uni.addInterceptor`方法，可以为特定的请求类型（如`request`）添加拦截器。
- **参数说明**：
  - 第一个参数是请求类型的字符串（如`'request'`）。
  - 第二个参数是一个对象，包含具体的拦截逻辑，比如`invoke`函数。

- **服务端信息**：需要在请求头中添加特定的字段，如`source-client`来标识请求来源。
- **请求参数**：可以在请求中添加必要的参数，例如`Authorization` token等。

### 4. 应用场景
- **统一管理**：通过拦截器，可以在一个地方集中管理所有请求的配置，提高代码的可维护性。
